Generative AI is a branch of artificial intelligence that focuses on creating new content such as text, images, audio, video, and even code by learning patterns from existing data. Overall, generative AI is transforming industries by improving efficiency, creativity, and personalization, although it also raises concerns related to ethics, bias, and misinformation.


```python
'''Question 2: Explain the role of probabilistic modeling in generative models. How do
these models differ from discriminative models?'''
```

Probabilistic modeling plays a central role in generative models by providing a mathematical framework to represent how data is generated. In generative modeling, the goal is to learn the **joint probability distribution** ( P(X, Y) ) or sometimes just ( P(X) ), where ( X ) represents the data and ( Y ) represents labels. By learning this distribution, the model can generate new, realistic data samples that resemble the training data. For example, models based on Bayesian inference use probabilities to capture uncertainty and variability in data. This probabilistic approach allows generative models to handle missing data, simulate new data points, and model complex real-world phenomena. Techniques such as Gaussian Mixture Models, Hidden Markov Models, and modern deep generative models like Variational Autoencoders (VAEs) and Generative Adversarial Networks (GANs) are all rooted in probabilistic principles.

Generative models differ fundamentally from discriminative models in their objectives and the type of probabilities they learn. Generative models learn the joint distribution ( P(X, Y) ) and can generate new data samples, making them useful for tasks like data synthesis, anomaly detection, and unsupervised learning. In contrast, discriminative models learn the **conditional probability** ( P(Y \mid X) ), directly focusing on predicting labels from given inputs without modeling how the data is generated. Algorithms like logistic regression, support vector machines, and neural network classifiers are examples of discriminative models. While discriminative models generally achieve higher performance in classification tasks due to their focus on decision boundaries, generative models offer greater flexibility by capturing the underlying data distribution, enabling them to generate new samples and work effectively even with incomplete data.


```python
'''Question 3: What is the difference between Autoencoders and Variational
Autoencoders (VAEs) in the context of text generation?'''
```

Autoencoders and Variational Autoencoders (VAEs) are both neural network architectures used for learning compact representations of data, but they differ significantly in how they model the latent space, especially in the context of text generation.

An **Autoencoder** consists of an encoder that compresses input text into a fixed latent vector and a decoder that reconstructs the original text from this representation. The goal is to minimize reconstruction error, meaning the output should be as close as possible to the input. However, the latent space learned by a standard autoencoder is **deterministic and unstructured**, which makes it difficult to generate new, meaningful text. If you sample random points from this latent space, the decoder often produces incoherent or invalid sentences because the model was not trained to organize the latent space in a smooth or continuous way.

In contrast, a **Variational Autoencoder (VAE)** introduces a probabilistic approach to latent representation using concepts from Bayesian inference. Instead of mapping input text to a single point, the encoder maps it to a **distribution** (usually a Gaussian defined by mean and variance). During training, the model samples from this distribution and uses a regularization term (KL divergence) to ensure that the latent space is smooth and continuous. This structured latent space allows VAEs to generate new text by sampling meaningful points, making them more suitable for text generation tasks.

In summary, the key difference is that **autoencoders learn a fixed encoding focused on reconstruction**, while **VAEs learn a probabilistic latent space that enables controlled and meaningful generation of new text**. As a result, VAEs are more powerful for generative tasks, whereas standard autoencoders are mainly useful for compression and feature extraction.


```python
'''Question 4: Describe the working of attention mechanisms in Neural Machine
Translation (NMT). Why are they critical?
'''
```

In Neural Machine Translation (NMT), attention mechanisms are used to help the model focus on the most relevant parts of the input sentence while generating each word of the output sentence. In a basic encoderвҖ“decoder model, the encoder converts the entire input sentence into a single fixed-length vector, which the decoder then uses to produce the translated sentence. However, this approach struggles with long or complex sentences because important information can be lost in that single vector.

The attention mechanism solves this problem by allowing the decoder to **look back at all the encoderвҖҷs hidden states** instead of relying on just one summary vector. At each step of translation, the decoder computes a set of attention weights that indicate how important each input word is for generating the current output word. These weights are calculated using alignment scores (e.g., dot product or feedforward networks) between the decoderвҖҷs current state and each encoder state. The model then creates a **context vector** as a weighted sum of the encoder hidden states, and this context is used along with the decoder state to predict the next word. This process is repeated for every word in the output sequence, dynamically shifting focus across the input sentence.

Attention mechanisms are critical in NMT for several reasons. First, they significantly improve translation quality, especially for long sentences, by preventing the loss of important information. Second, they enable better alignment between source and target languages, effectively learning which words correspond to each other during translation. Third, attention provides a level of interpretability, as the attention weights can show which input words the model is focusing on at each step. Finally, attention is the foundation of modern architectures like the Transformer, which relies entirely on attention (such as self-attention) instead of recurrent networks, leading to faster training and better performance.


```python
'''Question 5: What ethical considerations must be addressed when using generative AI
for creative content such as poetry or storytelling?
'''
```

When using generative AI for creative content such as poetry or storytelling, several important ethical considerations must be addressed to ensure responsible and fair use. One major concern is **authorship and originality**. Since generative AI systems like ChatGPT are trained on large datasets that may include copyrighted works, there is a risk that generated content may unintentionally resemble or replicate existing authorsвҖҷ styles or ideas, raising questions about intellectual property and ownership.

Another key issue is **plagiarism and copyright infringement**. AI-generated content might closely mimic specific writers or reproduce parts of original works without proper attribution, which can lead to legal and ethical problems. Closely related is the concern of **bias and fairness**. Generative AI models learn from historical data, which may contain social, cultural, or gender biases. As a result, the generated stories or poems might unintentionally reinforce stereotypes or exclude certain perspectives.

**Misinformation and misuse** is also a critical concern. AI can generate highly realistic narratives that may blur the line between fiction and reality, potentially misleading readers if not clearly identified as AI-generated. Additionally, there are concerns about **transparency**, where users should be informed when content is created by AI rather than a human author.

Another important consideration is the **impact on human creativity and employment**. As AI becomes more capable of producing high-quality creative content, it may affect writers, poets, and artists by reducing opportunities or devaluing human-created work. Finally, issues of **accountability** ariseвҖ”if AI-generated content is harmful, offensive, or misleading, it is important to determine who is responsible: the developer, the user, or the organization deploying the system.

In summary, ethical use of generative AI in creative fields requires attention to authorship, copyright, bias, transparency, misuse, and accountability, ensuring that the technology supports creativity without harming individuals or society.




```python
'''Question 6: Use the following small text dataset to train a simple Variational
Autoencoder (VAE) for text reconstruction:
["The sky is blue", "The sun is bright", "The grass is green",
"The night is dark", "The stars are shining"]
1. Preprocess the data (tokenize and pad the sequences).
2. Build a basic VAE model for text reconstruction.
3. Train the model and show how it reconstructs or generates similar sentences.
Include your code, explanation, and sample outputs.
'''
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# =========================
# 1. Dataset
# =========================
sentences = [
    "The sky is blue",
    "The sun is bright",
    "The grass is green",
    "The night is dark",
    "The stars are shining"
]

# =========================
# 2. Preprocessing
# =========================
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)

sequences = tokenizer.texts_to_sequences(sentences)

vocab_size = len(tokenizer.word_index) + 1
max_len = max(len(seq) for seq in sequences)

padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post')
padded_sequences = padded_sequences.astype("int32")

# =========================
# 3. Encoder
# =========================
embedding_dim = 16
latent_dim = 8

inputs = tf.keras.Input(shape=(max_len,))
x = layers.Embedding(vocab_size, embedding_dim)(inputs)
x = layers.Flatten()(x)
x = layers.Dense(16, activation='relu')(x)

z_mean = layers.Dense(latent_dim)(x)
z_log_var = layers.Dense(latent_dim)(x)

def sampling(args):
    z_mean, z_log_var = args
    epsilon = tf.random.normal(shape=(tf.shape(z_mean)[0], latent_dim))
    return z_mean + tf.exp(0.5 * z_log_var) * epsilon

z = layers.Lambda(sampling)([z_mean, z_log_var])

encoder = tf.keras.Model(inputs, [z_mean, z_log_var, z])

# =========================
# 4. Decoder
# =========================
latent_inputs = tf.keras.Input(shape=(latent_dim,))
x = layers.Dense(16, activation='relu')(latent_inputs)
x = layers.Dense(max_len * embedding_dim, activation='relu')(x)
x = layers.Reshape((max_len, embedding_dim))(x)

outputs = layers.TimeDistributed(
    layers.Dense(vocab_size, activation='softmax')
)(x)

decoder = tf.keras.Model(latent_inputs, outputs)

# =========================
# 5. VAE Model
# =========================
class VAE(tf.keras.Model):
    def __init__(self, encoder, decoder):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder

    def train_step(self, data):
        x = data[0] if isinstance(data, tuple) else data

        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(x)
            y_pred = self.decoder(z)

            # Reconstruction loss
            loss_fn = tf.keras.losses.SparseCategoricalCrossentropy()
            recon_loss = tf.reduce_mean(loss_fn(x, y_pred))

            # KL divergence
            kl_loss = -0.5 * tf.reduce_mean(
                1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var)
            )

            total_loss = recon_loss + kl_loss

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))

        return {
            "loss": total_loss,
            "reconstruction_loss": recon_loss,
            "kl_loss": kl_loss
        }

vae = VAE(encoder, decoder)
vae.compile(optimizer='adam')

# =========================
# 6. Training
# =========================
vae.fit(padded_sequences, epochs=100, batch_size=2, verbose=1)

# =========================
# 7. Reconstruction
# =========================
predictions = vae.decoder(vae.encoder(padded_sequences)[2]).numpy()

reverse_word_index = {v: k for k, v in tokenizer.word_index.items()}

def decode_sequence(seq):
    words = []
    for idx in seq:
        if idx != 0:
            words.append(reverse_word_index.get(idx, ""))
    return " ".join(words)

print("\n--- Reconstruction Results ---\n")
for i in range(len(predictions)):
    pred_seq = np.argmax(predictions[i], axis=1)
    print("Original      :", sentences[i])
    print("Reconstructed :", decode_sequence(pred_seq))
    print()
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# Small toy dataset
eng_sentences = [
    "hello", "how are you", "i am fine", "thank you", "good morning"
]

spa_sentences = [
    "hola", "como estas", "estoy bien", "gracias", "buenos dias"
]
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# Small toy dataset
eng_sentences = [
    "hello", "how are you", "i am fine", "thank you", "good morning"
]

spa_sentences = [
    "hola", "como estas", "estoy bien", "gracias", "buenos dias"
]
```

```python
# Tokenizers
eng_tokenizer = Tokenizer()
spa_tokenizer = Tokenizer()

eng_tokenizer.fit_on_texts(eng_sentences)
spa_tokenizer.fit_on_texts(spa_sentences)

eng_seq = eng_tokenizer.texts_to_sequences(eng_sentences)
spa_seq = spa_tokenizer.texts_to_sequences(spa_sentences)

# Padding
max_len_eng = max(len(seq) for seq in eng_seq)
max_len_spa = max(len(seq) for seq in spa_seq)

eng_seq = pad_sequences(eng_seq, maxlen=max_len_eng, padding='post')
spa_seq = pad_sequences(spa_seq, maxlen=max_len_spa, padding='post')

vocab_eng = len(eng_tokenizer.word_index) + 1
vocab_spa = len(spa_tokenizer.word_index) + 1
```

```python
embedding_dim = 32
units = 64

encoder_inputs = tf.keras.Input(shape=(max_len_eng,))
enc_emb = layers.Embedding(vocab_eng, embedding_dim)(encoder_inputs)

encoder_lstm = layers.LSTM(units, return_sequences=True, return_state=True)
encoder_outputs, state_h, state_c = encoder_lstm(enc_emb)
```

```python
# Decoder inputs
decoder_inputs = tf.keras.Input(shape=(max_len_spa,))
dec_emb = layers.Embedding(vocab_spa, embedding_dim)(decoder_inputs)

decoder_lstm = layers.LSTM(units, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(dec_emb, initial_state=[state_h, state_c])

# Attention layer
attention = layers.Attention()
context = attention([decoder_outputs, encoder_outputs])

# Combine context + decoder output
concat = layers.Concatenate()([decoder_outputs, context])
```

```python
dense = layers.Dense(vocab_spa, activation='softmax')
outputs = dense(concat)

model = tf.keras.Model([encoder_inputs, decoder_inputs], outputs)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy')

model.summary()
```

```python
# Shift target for teacher forcing
decoder_input_data = spa_seq[:, :-1]
decoder_target_data = spa_seq[:, 1:]

decoder_target_data = np.expand_dims(decoder_target_data, -1)
```

```python
model.fit(
    [eng_seq, decoder_input_data],
    decoder_target_data,
    epochs=200,
    verbose=1
)
```

```python
reverse_spa = {v: k for k, v in spa_tokenizer.word_index.items()}

def decode_sequence(input_seq):
    target_seq = np.zeros((1, max_len_spa))

    prediction = model.predict([input_seq, target_seq])
    pred_seq = np.argmax(prediction[0], axis=1)

    return " ".join([reverse_spa.get(i, "") for i in pred_seq if i != 0])

# Test
for i in range(len(eng_seq)):
    print("English:", eng_sentences[i])
    print("Spanish:", decode_sequence(eng_seq[i].reshape(1, -1)))
    print()
```
